```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x=np.array([5,15,25,35,45,55]).reshape((-1,1))
y=np.array([5,20,14,32,22,38])
```

```python
model=LinearRegression()
model.fit(x,y)

unknown_x=np.array([20]).reshape(1,-1)
print(model.predict(unknown_x))

plt.scatter(x,y)
plt.plot(x, model.coef_ * x + model.intercept_)
print(model.coef_)
print(model.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
x=np.array([5,10,15,20,65,70,75,80]).reshape((-1,1))
y=np.array([0,0,0,0,1,1,1,1])
model=LogisticRegression(solver='liblinear', random_state=0)#sag
model.fit(x,y)
unknown_x=np.array([20]).reshape(-1,1)
model.predict(unknown_x)

y_model=model.predict_proba(x)
y_model=np.delete(y_model,0,1)
plt.scatter(x,y)
plt.plot(x,y_model)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
data=pd.read_csv('Iris.csv', header=None, names=['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'])
data.drop( data[ data['class'] == 'Setosa' ].index, inplace=True)
x = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy()
encoder=OrdinalEncoder()
target=data['class'].unique().reshape(-1,1)

enc_y = encoder.fit_transform(target)
y = encoder.transform(data['class'].to_numpy().reshape(-1, 1))

model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(x, np.ravel(y))

print(data.head())

unknown_x = np.array([ 6.4, 3.2, 4.5, 1.5]).reshape(-1,4)
encoder.inverse_transform(model.predict(unknown_x).reshape(1, -1))

print(encoder.inverse_transform(model.predict(unknown_x).reshape(1,-1)))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv", header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
x = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy()
encoder = OrdinalEncoder()
target = data['class'].unique().reshape(-1, 1)

enc_y = encoder.fit_transform(target)
y = encoder.transform(data['class'].to_numpy().reshape(-1, 1))
model = DecisionTreeClassifier(criterion="gini")
model.fit(x, y)

print(data.head())

unknown_x = np.array([ 6.4, 3.2, 4.5, 1.5]).reshape(-1,4)
print(encoder.inverse_transform(model.predict(unknown_x).reshape(1, -1)))

plt.figure(figsize=(12,12)) 
tree.plot_tree(model, filled=True)
plt.show()
```
