## Cleanse Missing Values

```python
## Dropping missing values
import pandas
print(pandas.read_csv('stores.csv'))
data=pandas.read_csv('stores.csv')
data
```

```python
data.dropna()
```

```python
## Replacing Missing Values
from numpy import NaN
data.replace({NaN:0.00})
```

```python
## Replacing with a Scalar Value
## Slide 14
data.fillna(4.08)
```

```python
data.fillna(method='pad')
```

```python
data.fillna(method='bfill')
```
